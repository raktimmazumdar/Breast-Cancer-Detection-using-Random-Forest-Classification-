```python
######## Importing libraries____________________________________________________________________________________________________

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
######## Importing dataset______________________________________________________________________________________________________

dataset = pd.read_csv('data.csv')
```

```python
print(dataset.count(0))
```

```python
dataset.drop('Unnamed: 32', axis = 1, inplace = True)
print(dataset.count(0))
print(dataset.head())
X = dataset.iloc[:,2:32].values
Y = dataset.iloc[:,1]
```

```python
print(X)
```

```python
print(Y)
```

```python
######## Label Encoding for the Cancer type____________________________________________________________________________________

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
print(Y)
```

```python
######## splitting the dataset into training set and test set__________________

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
######## Feature Scaling_______________________________________________________

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 2:] = sc.fit_transform(x_train[:, 2:])
x_test[:, 2:] = sc.transform(x_test[:, 2:])
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
########### Implementing Random Forrest Classifire______________________________________________________________________________

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=50, criterion='gini', random_state = 0)
classifier.fit(x_train,y_train)
```

```python
########## Predicting a New Result______________________________________________________________________________________________

print(classifier.predict([[17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189
]]))
```

```python
######### Predicting the test set and comparing the results_____________________________________________________________________

y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
########## Confusion Matrix and Accuracy_________________________________________________________________________________________

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
